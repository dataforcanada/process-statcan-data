```bash
DATA_FOLDER=/data

source ../.env
```

# 1. Export Digital Boundary Files

```bash
output_folder="${DATA_FOLDER}/boundaries/output/2021/digital_boundary_files"
mkdir -p ${output_folder}
```

```bash
files=('country_2021'
       'grc_2021'
       'pr_2021'
       'er_2021'
       'car_2021'
       'cd_2021'
       'ccs_2021'
       'cma_2021'
       'csd_2021'
       'fed_2021_2013'
       'dpl_2021'
       'fsa_2021'
       'pop_ctr_2021'
       'ct_2021'
       'da_2021'
       'db_2021'
       'ada_2021'
       )
```

## Export as Parquet

```bash
for file in ${files[@]}
do
    output_file="${output_folder}/${file}.parquet"
    echo "Exporting silver.${file} table to ${output_file}" 
    ogr2ogr \
        -lco COMPRESSION="ZSTD" \
        -lco CREATOR="www.dataforcanada.org" \
        -lco WRITE_COVERING_BBOX="YES" \
        -lco SORT_BY_BBOX="YES" \
        -f Parquet \
        ${output_file} \
        "PG:host=db dbname=${POSTGRES_DB} user=${POSTGRES_USER} password=${POSTGRES_PASSWORD} port=5432" \
        "silver.${file}"
done
```

## Export as FlatGeobuf

```bash
for file in ${files[@]}
do
    output_file="${output_folder}/${file}.fgb"
    echo "Exporting ${file} table to ${output_file}" 
    ogr2ogr \
        -f FlatGeobuf \
        -lco TITLE="${file}" \
        ${output_file} \
        "PG:host=db dbname=${POSTGRES_DB} user=${POSTGRES_USER} password=${POSTGRES_PASSWORD} port=5432" \
        "silver.${file}" \
        -nln ${file}
done
```

## Export as File Geodatabase

```bash
for file in ${files[@]}
do
    output_file="${output_folder}/${file}.gdb"
    echo "Exporting ${file} table to ${output_file}" 
    ogr2ogr \
        -f OpenFileGDB \
        -lco TARGET_ARCGIS_VERSION="ARCGIS_PRO_3_2_OR_LATER" \
        ${output_file} \
        "PG:host=db dbname=${POSTGRES_DB} user=${POSTGRES_USER} password=${POSTGRES_PASSWORD} port=5432" \
        "silver.${file}" \
        -nln ${file}
done
```

### Zip File Geodatabases

```bash
cd ${output_folder}
for file in ${files[@]}
do
    output_file="${file}.gdb"
    echo "Zipping ${output_file}"
    zip -r "${output_file}.zip" ${output_file}
    rm -rf ${output_file}
done
```
